```python
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np

import chromedriver_autoinstaller
from selenium import webdriver
```

```python
chromedriver_autoinstaller.install()
browser = webdriver.Chrome()
```

```python
browser.get("https://www.webtoons.com/id/dailySchedule?weekday=MONDAY")
```

```python
soup = BeautifulSoup(browser.page_source, 'lxml')
```

```python
soup
```

```python
genre = soup.find_all('p', class_='genre')
jdul = soup.find_all('p', class_= 'subj')
nama_author = soup.find_all('p', class_='author')
like = soup.find_all('em', class_='grade_num')
```

```python
genr = []
for each_genre in genre:
    genr.append(each_genre.text)
```

```python
judul = []
for each in jdul:
    judul.append(each.text)
```

```python
author = []
for each in nama_author:
    author.append(each.text)
```

```python
data ={
    'judul' : judul,
    'author' : author,
    'genre' : genr
}
```

```python
pd.DataFrame(data)
```

```python
pd.DataFrame(data).to_excel('scrapedata.xlsx')
```
